### Bubble Sort

Bubble Sort is one of the first ever created sorting algorithms, used for the first time in 1962.

It is called "Bubble Sort" because the bigger numbers go up and the smaller go down just like bubbles.

This algorithm uses a brute force (Complete search) approach for solving the sorting problem, because, we compare each number with all the array elements until is sorted, checking it for being the greater in the non-ordered subset of numbers.

### Implementation and Algorithm Analysis with RAM Model (Using "swapped" flag).

```python
import random

# Worst case || Average Case || Best case

def bubbleSort(arr):
    n = len(arr) # 3 || 3
    i = 0 # 3 || 3
    while (i < n): # 4(n+1) || 4(n+1) || 4(n+1)
        swapped = False # 3n || 3n
        j = 0 # 3n || 3n
        while j < n-i-1: # 7n (n/2) || 7n (n/2)
            if arr[j]>arr[j+1]: # 5n (n/2) || 5n (n/2)/2
                arr[j],arr[j+1] = arr[j+1],arr[j] #6n (n/2)|| 6n (n/2)/2
                swapped = True # 3n (n/2)|| 3n (n/2)/2
            j+=1 #3n (n/2) || 3n (n/2)/2
        if not swapped: # 3n || 3n/2 || 3
            break # 1
        i+=1 # 3n
    return arr # 1

#

# Worst case polynomial = 12 n^2 + 16 n + 12 = O(n^2)
# Average case polynomial = 17 n^2/4 + 13n + 8 = Theta(n^2)
# Best case polynomial (Only counting the linear for because of swapped flag) = 13n + 15 = Omega(n)



if __name__ == "__main__":
    arr = []
    for i in range(50):
        arr.append(random.randint(-500,500))
    print("Original: ",arr)
    sorted_arr = bubbleSort(arr)
    print("Sorted: ",sorted_arr)
```

### Merge Sort

Merge sort is one of the most efficient algorithms for sorting, created by John von Neumann in 1945.

It's called "Merge Sort" because the algorithm divides the array until gets "subsets" of a single element, then merge all the elements comparing between the subarrays until the array is sorted.

This algorithm uses a Divide and Conquer approach because divides the arrays until gets single values in logarithm time for increase the speed for rebuilding the array but now sorted in linear time.

### Implementation and Algorithm Analysis with RAM Model.

```python
import random

def mergeSort(arr):
    if len(arr) <= 1:# 4 | 1
        return arr # 1 | 1
    
    mid = len(arr) // 2 # 5 and being n our array size (The array length divides by two n/2) | 1
    left = arr[:mid] # Creating left subarray | n/2 | n/2
    right = arr[mid:] # Creating  right subarray n/4 | n/2
    
    left = mergeSort(left) # Recursive call (constant division //2 ) | log (n) | log(n)
    right = mergeSort(right)  # Recursive call (constant division //2 ) | 2 log (n) | log(n)
    
    return merge(left, right) # 60n + 24 * (2 log n + 10)

def merge(left, right):
    result = [] # 3 | 1
    i = 0 # 3 | 1
    j = 0 # 3 | 1
    
    while i < len(left) and j < len(right): # 7(2n+2) | 0
        if left[i] < right[j]: # 4(2n) | 1
            result.append(left[i]) # 4(n) | n | Es n porque 2n = left + right
            i += 1 # 4(n) | 1
        else:
            result.append(right[j]) # 4(n) | 1
            j += 1 # 4(n) | 1
    
    while i < len(left): # 5(n) | 0
        result.append(left[i]) # 3(n) | n
        i += 1 # 3(n) | n
    
    while j < len(right): # 5(n) | 0
        result.append(right[j]) # 3(n) | n
        j += 1 # 3(n) | n
    
    return result # 1 | 1

if __name__ == "__main__":
    arr = []
    for i in range(50):
        arr.append(random.randint(-500,500))
    print("Original: ",arr)
    sorted_arr = mergeSort(arr)
    print("Sorted: ",sorted_arr)
    
    
# Polynomial Merge 60n + 24 = O(n)
# Polynomial MergeSort 60n + 24 * (2 log n + 10) = O(n log n)
```

#### We can prove that all cases (Best, Average and Worst) are the same, because there is no condition to not divide the array and merging it again, and comparisons are the same, because all are if-else statements with the same operations. All cases are n log n

#### Plotting the polynomials:

```python
import numpy as np
import matplotlib.pyplot as plt
import random
times = 0
def mergeSort(arr):
    global times
    if len(arr) <= 1:
        return arr
    
    mid = len(arr) // 2
    left = arr[:mid]
    right = arr[mid:]
    
    left = mergeSort(left)
    right = mergeSort(right)
    times += 1
    return merge(left, right)

def merge(left, right):
    global times
    result = []
    i = 0
    j = 0
    
    while i < len(left) and j < len(right):
        times += 1
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    
    while i < len(left):
        times += 1
        result.append(left[i])
        i += 1
    
    while j < len(right):
        times += 1
        result.append(right[j])
        j += 1
    
    return result


n = np.linspace(1, 100, 100) 
y = (60*n + 24) * (2 * np.log2(n) + 10)
    
plt.plot(n, y)
plt.title('All cases theoretical complexity: ')
plt.show()

x = []
y = []
for tam in range(1,500):
    l = []
    x.append(tam)
    for value in range(tam):
        l.append(value)
    l = mergeSort(l)
    y.append(times)
    times = 0
        
plt.plot(x,y)
plt.title('Best case performance')
plt.grid(True)
plt.show()
    
x = []
y = []
for tam in range(1,500):
    l = []
    x.append(tam)
    for value in range(tam):
        l.append(random.randint(-500,500))
    l = mergeSort(l)
    y.append(times)
    times = 0
        
plt.plot(x,y)
plt.title('Average case performance')
plt.grid(True)
plt.show()
    
x = []
y = []
for tam in range(1,500):
    l = []
    x.append(tam)
    for value in range(tam):
        l.append(value)
        l.reverse()
    l = mergeSort(l)
    y.append(times)
    times = 0
        
plt.plot(x,y)
plt.title('Worst case performance')
plt.grid(True)
plt.show()
```
