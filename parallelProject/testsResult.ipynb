```python
def read_file(file_path):
    real_part = []
    imag_part = []

    with open(file_path, 'r') as file:
        next(file)
        for line in file:
            real_str, imag_str = line.strip().split(',')
  
            real_part.append(float(real_str))
            imag_part.append(float(imag_str))

    return real_part, imag_part


file_path = 'resultsSerial.txt'
serialr,seriali = read_file(file_path)
file_path = 'resultsParallel.txt'
parallelr,paralleli = read_file(file_path)
file_path = 'resultsTask.txt'
taskr, taski = read_file(file_path)


for i in range (len(serialr)):
    if serialr[i] != parallelr[i] or serialr[i] != taskr[i] or parallelr[i] != taskr[i]:
        print("Data inconsisency at term "+str(i))
        print(seriali[i])
        print(paralleli[i])
        print(taski[i])
        
for i in range (len(seriali)):
    if seriali[i] != paralleli[i] or seriali[i] != taski[i] or paralleli[i] != taski[i]:
        print("Data inconsisency at term "+str(i))
        print(seriali[i])
        print(paralleli[i])
        print(taski[i])
```

#### Performance graphs

```python
import matplotlib.pyplot as plt

def read_data(file_path):
    degrees = []
    times = []

    with open(file_path, 'r') as file:
        next(file)

        for line in file:
            degree, time = map(float, line.strip().split(','))
            
            degrees.append(degree)
            times.append(time)

    return degrees, times

file_path = 'serialTime.csv'
degreesS, timesS = read_data(file_path)

file_path = 'parallelTime.csv'
degreesP, timesP = read_data(file_path)

file_path = 'taskTime.csv'
degreesT, timesT = read_data(file_path)

plt.title("FFT performance Ryzen 9 6900 HS (8C/12T)")
plt.plot(degreesS,timesS, label="Serial performance")
plt.plot(degreesS,timesP,label="Parallel performance")  
plt.plot(degreesS,timesT,label="Task performance")  
plt.grid(True)
plt.legend()
plt.show()

start_degree = 2000
end_degree = 10000

selected_degreesS = degreesS[start_degree:end_degree]
selected_timesS = timesS[start_degree:end_degree]

selected_degreesP = degreesP[start_degree:end_degree]
selected_timesP = timesP[start_degree:end_degree]

selected_degreesT = degreesT[start_degree:end_degree]
selected_timesT = timesT[start_degree:end_degree]

plt.title("FFT performance - Zoom for finding the balance point (2k-10k)")
plt.plot(selected_degreesS, selected_timesS, label="Serial performance")
plt.plot(selected_degreesP, selected_timesP, label="Parallel performance")  
plt.plot(selected_degreesT, selected_timesT, label="Task performance")  
plt.grid(True)
plt.legend()
plt.show()

start_degree = 8192
end_degree = 16384

selected_degreesS = degreesS[start_degree:end_degree]
selected_timesS = timesS[start_degree:end_degree]

selected_degreesP = degreesP[start_degree:end_degree]
selected_timesP = timesP[start_degree:end_degree]

selected_degreesT = degreesT[start_degree:end_degree]
selected_timesT = timesT[start_degree:end_degree]

plt.title("FFT performance - Zoom for finding the balance point (8192-16384)")
plt.plot(selected_degreesS, selected_timesS, label="Serial performance")
plt.plot(selected_degreesP, selected_timesP, label="Parallel performance")  
plt.plot(selected_degreesT, selected_timesT, label="Task performance")  
plt.grid(True)
plt.legend()
plt.show()

start_degree = 4096
end_degree = 8192

selected_degreesS = degreesS[start_degree:end_degree]
selected_timesS = timesS[start_degree:end_degree]

selected_degreesP = degreesP[start_degree:end_degree]
selected_timesP = timesP[start_degree:end_degree]

selected_degreesT = degreesT[start_degree:end_degree]
selected_timesT = timesT[start_degree:end_degree]

plt.title("FFT performance - Zoom for finding the balance point (4096-8192)")
plt.plot(selected_degreesS, selected_timesS, label="Serial performance")
plt.plot(selected_degreesP, selected_timesP, label="Parallel performance")  
plt.plot(selected_degreesT, selected_timesT, label="Task performance")  
plt.grid(True)
plt.legend()
plt.show()
```
